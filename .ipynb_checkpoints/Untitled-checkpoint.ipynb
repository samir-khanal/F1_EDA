```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
circuits= pd.read_csv("data/circuits.csv")
constructor_result= pd.read_csv("data/constructor_results.csv")
constructor_standing= pd.read_csv("data/constructor_standings.csv")
constructors=pd.read_csv("data/constructors.csv")
driver_standings=pd.read_csv("data/driver_standings.csv")
drivers=pd.read_csv("data/drivers.csv")
lap_times=pd.read_csv("data/lap_times.csv")
pit_stops=pd.read_csv("data/pit_stops.csv")
qualifying=pd.read_csv("data/qualifying.csv")
races=pd.read_csv("data/races.csv")
results=pd.read_csv("data/results.csv")
seasons=pd.read_csv("data/seasons.csv")
sprint_results=pd.read_csv("data/sprint_results.csv")
status=pd.read_csv("data/status.csv")
```

```python
results.head()
races.head()
```

```python
# Merging results with races to get the date of each race
results_races = pd.merge(results, races[['raceId', 'date']], on='raceId', how='left')
#results_races.head()
```

```python
# Merging with drivers file to attach driver names to each result
results_races = pd.merge(results_races, drivers[['driverId', 'forename', 'surname']], on='driverId', how='left')

# Making full driver name column
results_races['Driver'] = results_races['forename'] + " " + results_races['surname']

# Converting the date to datetime
results_races['date'] = pd.to_datetime(results_races['date'])
results_races['year'] = results_races['date'].dt.year
results_races.head()
```

```python
# Filter for race wins (where positionOrder equals 1)
wins = results_races[results_races['positionOrder'] == 1]
wins.head(1110)
#wins.shape
```

```python

# Group wins by Driver and year to see how many wins per year each driver had
driver_wins = wins.groupby(['Driver', 'year']).size().reset_index(name='wins')

# Calculate total wins per driver and pick the top 5 drivers
total_wins = driver_wins.groupby('Driver')['wins'].sum().reset_index().sort_values(by='wins', ascending=False)
top5_drivers = total_wins.head(5)['Driver'].tolist()

# Filter data for top 5 drivers
top5_wins = driver_wins[driver_wins['Driver'].isin(top5_drivers)]

# Plot wins over the years for the top 5 drivers
plt.figure(figsize=(12, 6))
sns.lineplot(data=top5_wins, x='year', y='wins', hue='Driver', marker='o')
plt.title("Driver Wins Over the Years (Top 5 Drivers)")
plt.xlabel("Year")
plt.ylabel("Number of Wins")
plt.legend(title="Driver")
plt.show()
```
