```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
circuits= pd.read_csv("data/circuits.csv")
constructor_result= pd.read_csv("data/constructor_results.csv")
constructor_standing= pd.read_csv("data/constructor_standings.csv")
constructors=pd.read_csv("data/constructors.csv")
driver_standings=pd.read_csv("data/driver_standings.csv")
drivers=pd.read_csv("data/drivers.csv")
lap_times=pd.read_csv("data/lap_times.csv")
pit_stops=pd.read_csv("data/pit_stops.csv")
qualifying=pd.read_csv("data/qualifying.csv")
races=pd.read_csv("data/races.csv")
results=pd.read_csv("data/results.csv")
seasons=pd.read_csv("data/seasons.csv")
sprint_results=pd.read_csv("data/sprint_results.csv")
status=pd.read_csv("data/status.csv")
```

```python
#*******   Driver Performance Over the Years   **********
results.head()
races.head()
```

```python
# Merging results with races to get the date of each race
results_races = pd.merge(results, races[['raceId', 'date']], on='raceId', how='left')
#results_races.head()
```

```python
# Merging with drivers file to attach driver names to each result
results_races = pd.merge(results_races, drivers[['driverId', 'forename', 'surname']], on='driverId', how='left')

# Making full driver name column
results_races['Driver_Name'] = results_races['forename'] + " " + results_races['surname']

# Converting date to datetime
results_races['date'] = pd.to_datetime(results_races['date'])
results_races['year'] = results_races['date'].dt.year
results_races.head()
#print(results_races.isnull().sum())
```

```python
# Filtering the race wins (by using positionOrder equals 1)
wins = results_races[results_races['positionOrder'] == 1]
wins.head()
# year_to_filter = 1954
# filtered_data = results_races[results_races['year'] == year_to_filter]
# print(filtered_data)
#wins.shape
```

```python
# Grouping wins by Drivers name and year to see how many wins per year each driver had
driver_wins = wins.groupby(['Driver_Name', 'year']).size().reset_index(name='wins')
print(driver_wins)

# finding total wins per driver and choosing the best top 5 drivers
total_wins = driver_wins.groupby('Driver_Name')['wins'].sum().reset_index().sort_values(by='wins', ascending=False)
top5_drivers = total_wins.head(5)['Driver_Name'].tolist()
print(top5_drivers)

# Filtering data for top 5 drivers
top5_wins = driver_wins[driver_wins['Driver_Name'].isin(top5_drivers)]
print(top5_wins)
#print(top5_wins.isnull().sum())
```

```python

# Plot wins over the years for the top 5 drivers
plt.figure(figsize=(12, 6))
sns.lineplot(data=top5_wins, x='year', y='wins', hue='Driver_Name', marker='o')
plt.title("Driver Wins Over the Years (Top 5 Drivers)")
plt.xlabel("Year")
plt.ylabel("Number of Wins")
plt.legend(title="Driver")
plt.show()
```

```python
#******  Reason for finding drivers performance over the years is:
#******  1) To understand which drivers have dominated different eras and how their performance evolved over time
#******  2) Visualizing wins over time helps reveal trends and milestones in F1 history
```

```python
# Merging results with constructors to link each race result with a team
results_constructors = pd.merge(results, constructors[['constructorId', 'name']], on='constructorId')
print(results_constructors)

constructor_df=pd.merge(constructors,constructor_standing, on = "constructorId")
print(constructor_df)
```

```python
constructor_points = constructor_df.groupby('name')['points'].sum()
constructor_points
```

```python
# Plot total wins per constructor
plt.figure(figsize=(10, 6))
sns.barplot(data=top10_teams, x='constructor_wins', y='name', palette='viridis')
plt.title("Top 10 Constructor based on wins")
plt.xlabel("Number of Wins")
plt.ylabel("Constructor")
plt.show()
```

```python
#***********       Impact of Pole Position on Race Wins      ************
```

```python
# Loading qualifying data as it contains information on qualifying positions
qualifying = pd.read_csv('data/qualifying.csv')
qualifying.head()
```

```python
# Filtering the qualifying data to include only pole positions (position == 1)
pole_positions = qualifying[qualifying['position'] == 1][['raceId', 'driverId']]
pole_positions.head()
```

```python
# Merge pole position data with race results to check if the pole sitter won
results_with_pole = pd.merge(results, pole_positions, on=['raceId', 'driverId'], how='inner') #how='inner', Ensures to only keep races where a driver started on pole.
results_with_pole.head()
```

```python
# Now filter to see how many pole sitters won the race
wins_from_pole = results_with_pole[results_with_pole['positionOrder'] == 1]
total_poles = pole_positions.shape[0]  #total number of pole positions 
win_rate = wins_from_pole.shape[0] / total_poles * 100   #Total number of wins from pole.
print(f"total no of poles:{total_poles}")
print(f"Win rate from pole position: {win_rate:.2f}%")
```

```python
# Visualize the distribution of finishing positions for pole sitters
plt.figure(figsize=(10, 6))
sns.countplot(x='positionOrder', data=results_with_pole, palette='magma')
plt.title("Finish Positions of Pole Sitters")
plt.xlabel("Finish Position")
plt.ylabel("Count")
plt.show()
```

```python
# Data for the pie chart
labels = ['Wins from Pole', 'Did not Win']
sizes = [wins_from_pole.shape[0], total_poles - wins_from_pole.shape[0]]

# Create pie chart
plt.figure(figsize=(5, 5))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Percentage of Wins from Pole Position")
plt.show()
```

```python
#******  Reason for visualizing Impact of Pole Position on Race Wins is:
#******  1)To analyze how often a driver who starts in the pole position goes on to win the race
#******  2)This reveals the importance of qualifying performance and the effect of race strategy
```
